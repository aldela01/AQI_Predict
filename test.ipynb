```python
import pandas as pd
import re
```

```python
# Remove max columns limit for display
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Solicitud_Historica/Historico_meteo.csv", sep=";",
                )
```

```python
# Get all columns that start with 'CAL-LASA'
columns = ['est'] + [col for col in df.columns if col.startswith('GIR-SOSN')]
# Extract the date from the column names
df_gir_sosn=df[columns]
```

```python
df_gir_sosn.head()
```

```python
df_gir_sosn.columns = df_gir_sosn.iloc[0]
```

```python
# Rename nan columns to 'Fecha_Hora'
df_gir_sosn = df_gir_sosn.rename(columns=lambda x: 'Fecha_Hora' if pd.isna(x) else x)
```

```python
df_gir_sosn.head()
```

```python
df_gir_sosn.drop(index=0, inplace=True)
```

```python
# Convertir la columna 'Fecha_Hora' a tipo datetime
df_gir_sosn['Fecha_Hora'] = pd.to_datetime(df_gir_sosn['Fecha_Hora'], format='%d/%m/%Y %H:%M')
```

```python
# Get only the rows with 'Fecha_Hora' after 2018-01-01
df_gir_sosn = df_gir_sosn[df_gir_sosn['Fecha_Hora'] > '2018-01-01'].reset_index(drop=True)
```

```python
df_gir_sosn['DViento_SSR'] = df_gir_sosn['DViento_SSR'].str.replace(',', '.').astype(float)
```

```python
import pandas as pd
```

```python
pm25_df = pd.read_csv('Solicitud_Historica/Historico_pm25.csv', sep=';')
```

```python
pm25_df.shape
```

```python
pm25_df.head()
```

```python
# Rename the first column to 'fecha_hora'
pm25_df.rename(columns={pm25_df.columns[0]: 'fecha_hora'}, inplace=True)
```

```python
pm25_df['fecha_hora'] = pd.to_datetime(pm25_df['fecha_hora'], format='%d/%m/%y %H:%M')
```

```python
pm25_df_bar_torr = pm25_df[['fecha_hora', 'BAR-TORR']].copy()
```

```python
pm25_df_bar_torr.head()
```

```python
# Get the data after 2018-01-01
pm25_df_bar_torr = pm25_df_bar_torr[pm25_df_bar_torr['fecha_hora'] >= '2018-01-01']
```

```python
pm25_df_bar_torr.info()
```

```python
pm25_df_bar_torr.tail()
```

```python
# Export the DataFrame to a CSV file
pm25_df_bar_torr.to_csv('data/SIATA/82_pm25_hr.csv', index=False)
```

# Concatenar PM2.5 y variables meteorológicas

```python
meteo_df = pd.read_csv('data/SIATA/82_meteo_hr.csv', parse_dates=['fecha_hora'])
```

```python
meteo_df.head()
```

```python
# Concatenate the PM2.5 data with the meteorological data
pm25_meteo_df = pd.merge(meteo_df, pm25_df_bar_torr, on='fecha_hora', how='outer')
```

```python
pm25_meteo_df.shape
```

```python
# Rename the BAR-TORR column to PM25
pm25_meteo_df.rename(columns={'BAR-TORR': 'PM25'}, inplace=True)
```

```python
# Export the DataFrame to a CSV file
pm25_meteo_df.to_csv('data/SIATA/82.csv', index=False)
```

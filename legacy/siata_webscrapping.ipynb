```python
import pandas as pd
import requests
```

```python
def load_initial_page():
    url = "https://siata.gov.co/descarga_siata/index.php/index2/login/cerrar_sesion"
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        raise Exception(f"Failed to load page: {response.status_code}")
```

```python
def get_report():
    url = "https://siata.gov.co/descarga_siata/application/assets/assets-siata/downloads/MgChVco4aGiI6SyGTCx4Rl/estacion_data_humedad_419__20190901_20190930.csv"
    response = requests.get(url)
    if response.status_code == 200:
        return response
    else:
        raise Exception(f"Failed to load page: {response.status_code}")
```
